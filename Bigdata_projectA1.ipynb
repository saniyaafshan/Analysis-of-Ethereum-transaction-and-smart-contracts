```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/content/output_partA.csv')
df.columns = ['Date', 'Transactions per month']
```

```python
df
```

```python
import matplotlib.pyplot as plt

x = df['Date']
y = df['Transactions per month']

plt.figure(figsize=(12, 8)) 

plt.xticks(rotation=90)
plt.bar(x, y, color='green') 
plt.title('Transactions per month')
plt.xlabel('Months')
plt.ylabel('Transactions')

plt.grid() 
plt.legend(['Transactions']) 

plt.show()
```
